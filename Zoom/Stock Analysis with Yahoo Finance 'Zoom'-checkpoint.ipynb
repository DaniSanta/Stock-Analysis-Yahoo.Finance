```python
#Import libraries
import pandas as pd 
import matplotlib.pyplot as plt 
import pandas_datareader as dr 
import numpy as np 
```

```python
#dr.data.get_data_yahoo/Information from yahoo.finance 
df = dr.data.get_data_yahoo('zm',start='2019-04-18', end='2020-03-24')
df.head()
```

```python
#Selecting Close
df_close = pd.DataFrame(df.Close)
```

```python
#MA in order to identify the trend direction and to determine support and resistance levels.
df_close['MA_9'] = df_close.Close.rolling(9).mean().shift()
df_close['MA_21'] = df_close.Close.rolling(21).mean()
```

```python
df_close['MA_9'].head(10)
```

```python
#Graph showing the 
plt.figure(figsize=(15,10))
plt.grid(True)
plt.plot(df_close['Close'],label='ZM')
plt.plot(df_close['MA_9'],label='MA 9 day')
plt.plot(df_close['MA_21'],label='MA 21 day')
plt.legend(loc=2)
```

```python
df_close["Change"] = np.log(df_close["Close"]/df_close['Close'].shift())
```

```python
plt.plot(df_close.Change)
```

```python
#calculate the rolling std daily price change  
df_close['Volatility'] = df_close.Change.rolling(21).std().shift()
df_close['Volatility'].plot(figsize=(15,5))
```

```python
#Compared the Close & Open price of 'Zoom' shares since its IPO 
df[['Open','Close']].plot(figsize=(15,5))
```

```python
#Percent change for 'zoom' share closing price 
df['Close'].pct_change().plot(figsize=(15,5))
```

```python
#Offset
df_close['Actual_Change'] = df_close['Close'] - df_close['Close'].shift(1)
```

```python
df_close['Expected_Change'] = df_close['Close'].shift(1) * df_close['Volatility']
```

```python
df_close.head()
```

```python
#Running more than once will continosly give a smaller data set / cell , all output , clear to reset.
df_close = df_close.iloc[22:]
df_close.head()
```

```python
#The magnitude in std 
df_close['Magnitude'] = df_close['Actual_Change'] / df_close['Expected_Change']
```

```python
#Histogram of the magnitude in std 
plt.hist(df_close['Magnitude'])
```
